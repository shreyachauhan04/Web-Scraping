```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.amazon.in/OnePlus-Green-Woods-256GB-Storage/product-reviews/B097RF4WCM/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
code = requests.get(url)
code
```

```python
soup = BeautifulSoup(code.content,'html.parser')
print(soup)
```

```python
names = soup.select('span.a-profile-name')[2:]
names
```

```python
len(names)
```

```python
titles = soup.select('a.review-title span')
titles
```

```python
len(titles)
```

```python
stars = soup.select('i.review-rating span.a-icon-alt')[2:]
stars
```

```python
len(stars)
```

```python
dates = soup.select('span.review-date')[2:]
dates
```

```python
len(dates)
```

```python
reviews = soup.select('span.review-text-content span')
reviews
```

```python
len(reviews)
```

```python
cust_name = list()
rev_date = list()
rev_title = list()
ratings = list()
content = list()
for i in range(len(names)):
  cust_name.append(names[i].get_text())
  rev_date.append(dates[i].get_text().replace("Reviewed in India on ",""))
  rev_title.append(titles[i].get_text())
  ratings.append(stars[i].get_text().replace(" out of 5 stars",""))
  content.append(reviews[i].get_text().strip("\n "))
```

```python
import pandas as pd
df = pd.DataFrame()
df['Date'] = rev_date
df['Customer Name'] = cust_name
df['Ratings'] = ratings
df['Title'] = rev_title
df['Reviews'] = content
df
```

```python
df.to_csv("reviews.csv")
```

```python
cust_name = list()
rev_date = list()
rev_title = list()
ratings = list()
content = list()
error = list()
for page in range(1,51):
  url = "https://www.amazon.in/11X-Celestial-Silver-128GB-Storage/product-reviews/B085J1DRPF/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber="+str(page)
  code = requests.get(url)
  if str(code) == "<Response [200]>":
    soup = BeautifulSoup(code.content,'html.parser')
    names = soup.select('span.a-profile-name')[2:]
    titles = soup.select('a.review-title span')
    dates = soup.select('span.review-date')[2:]
    reviews = soup.select('span.review-text-content span')
    stars = soup.select('i.review-rating span.a-icon-alt')[2:]
    for i in range(len(names)):
      cust_name.append(names[i].get_text())
      rev_date.append(dates[i].get_text().replace("Reviewed in India on ",""))
      rev_title.append(titles[i].get_text())
      ratings.append(stars[i].get_text().replace(" out of 5 stars",""))
      content.append(reviews[i].get_text().strip("\n "))
  else:
    error.append(page)
```

```python
len(error)
```

```python
import pandas as pd
df = pd.DataFrame()
df['Date'] = rev_date
df['Customer Name'] = cust_name
df['Ratings'] = ratings
df['Title'] = rev_title
df['Reviews'] = content
df
```
